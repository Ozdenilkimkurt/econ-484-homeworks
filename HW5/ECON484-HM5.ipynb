```
import numpy as np
import pandas as pd
```

```
#CSV file is loaded. Data has been manupulated by separating semicolons.
dataset = pd.read_csv('Econ484-Spring2022-HW5-Dataset.csv', sep=';')
```

```
dataset
```

```
#Dataset "nan" values has been replaced with zero.
data = dataset.fillna(0)
```

```
data
```

```
'''The main focus of ICA is exact source of the given context. For example, if there are two microphones, one of them just captures
the red voices, the other one captures the blue ones. The voice has been measured but we need to find the exact colors of the voices,
ICA helps for that in a noisy data. 
On the other hand, PCA fails at recovering colors because all signals reflect non-Gaussian processes. For our example, the captured 
voices of blue and red are harmonized and gives us the mixed color like purple. 
For this project, I only choice HM1, MT, PROJ and FINAL columns. Because quarters were only given too similar (hard the recognizable
data within noise. So for my purpose gathering "MT AND FINAL" against "HW1 AND PROJ" was meainingful. Because HM1 and PROJ had
contrary notes while MT and FINAL had similar notes.'''

similar1 = np.c_[data['HW1'],data['PROJ']]
similar2 = np.c_[data['MT'],data['FINAL']]

# 0.2 Noise added
similar1 += 0.2 * np.random.normal(size=similar1.shape)  
similar2 += 0.2 * np.random.normal(size=similar2.shape)  
```

```
similar1
```

```
similar2
```

```
# Standardized the data
similar1 /= similar1.std(axis=0)  
similar2 /= similar2.std(axis=0) 
```

```
similar1
```

```
similar2
```

```
np.asarray(similar1)
np.asarray(similar2)
X = np.dot(similar1, similar2.T)
```

```
from sklearn.decomposition import FastICA, PCA

# ICA has been applied with 3 n components.
ica = FastICA(n_components=3)
#Signals have been reconstructed.
icaFit = ica.fit_transform(X)  
#Estimated mix matrix achieved.
Mixed = ica.mixing_ 
```

```

# In order to make comparison with ICA, PCA is estimated.
pca = PCA(n_components=3)
#Signals have been reconstructed vertically.
pcaFit = pca.fit_transform(X)
```

```
import matplotlib.pyplot as plt

plt.figure()

models = [X, similar1, icaFit, pcaFit]
names = [
    "Observations (mixed signal)",
    "True Sources",
    "ICA recovered signals",
    "PCA recovered signals",
]
colors = ["red", "blue", "green"]
plt.figure(figsize=(12,7))
#For loop for distributing to models and graph names.
for i, (model, name) in enumerate(zip(models, names), 1):
    plt.subplot(4, 1, i)
    
    plt.title(name)
#Giving colors to model
    for signal, color in zip(model.T, colors):
        plt.plot(signal, color=color)

#A tight graph is applied.

plt.tight_layout()

plt.show()

''' As we can see, ICA results are more precise given exact measures. But PCA results only shows weak signals.
Part 4: For considering our course ECON484, I think the better measurement is obtained by Independent Component Analyses,
because it allows us to get precise information in a noisy data about the main sources.'''
```
